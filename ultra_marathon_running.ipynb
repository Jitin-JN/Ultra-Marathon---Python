Get distances where distance is either 50km or 50mi for the year 2020 where country is USA

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine both the filters

df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.shape
```

Remove USA from events name


```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

Clean up athelete age

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#Remove h from athelete performance

df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# Drop columns - Athlete Club, Athlete Country, Athlete year of birth, Athelete age category

df2.drop(['Athlete club','Athlete age category','Athlete year of birth'],axis = 1, inplace = True)

```

```python
df2
```

Clean Up Null Values

```python
df2.isnull().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#Check duplicates

df2[df2.duplicated() == True]
```

```python
df2.reset_index(drop = True)
```

Fix Data Types

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.head()
```

Rename Columns

```python
df2.columns
```

```python
df2 = df2.rename(columns= {'Year of event' : 'year',
                           'Event dates' : 'race_day',
                           'Event name' : 'race_name',
                           'Event distance/length' : 'race_length',
                           'Event number of finishers' : 'race_number_of_finishers',
                           'Athlete performance' : 'athlete_performance',
                           'Athlete country' : 'athlete_country',
                           'Athlete gender' : 'athlete_gender',
                           'Athlete average speed' : 'athlete_average_speed',
                           'Athlete ID' : 'athlete_id'})
```

Find difference in speed for 50 km and 50 mile male to female

```python
df2.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

Find which age groups are best in the 50mile race, should have completed minimum of 20 races

```python
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19')
```

Find which age groups are worst in the 50mile race , should have completed minimum of 20 races

```python
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>19')
```

Seasons for the data -> slower in summer than winter?

```python
df2['race_month']= df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df2.tail(25)
```

```python
df2.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by ='mean', ascending = False)
```

```python
df2.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by ='mean', ascending = False)
```
